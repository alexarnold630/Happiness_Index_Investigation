```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath1 = "Resources/Regions_2020.csv"

df_complete_2020 = pd.read_csv(filepath1)
df_complete_2020.head()
```

```python
df_complete_2020.describe()
```

```python
#histograms
for col in df_complete_2020.columns:
    
    plt.figure(figsize=(10,6))
    plt.hist(df_complete_2020[col])
    plt.title(col)
    plt.ylabel("Count")
    plt.show()
    
    print()
```

```python
 #boxplots
for col in df_complete_2020.columns:
    
    plt.figure(figsize=(10,6))
    plt.boxplot(df_complete_2020[col], showmeans=True)
    plt.title(col)
    plt.ylabel("Count")
    plt.show()
    
    print()
```

```python
corrs = df_complete_2020.corr()
corrs
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corrs)
plt.show()
```
