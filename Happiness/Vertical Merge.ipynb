```python
# data science imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# linear modelling
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import scipy.stats as sts
```

```python
filepath1 = "Clean_2015.csv"
filepath2 = "Clean_2016.csv"
filepath3 = "Clean_2017.csv"
filepath4 = "Clean_2018.csv"
filepath5 = "Clean_2019.csv"
filepath6 = "Clean_2020.csv"

df_2015 = pd.read_csv(filepath1)
df_2015['Year'] = 2015
df_2016 = pd.read_csv(filepath2)
df_2016['Year'] = 2016
df_2017 = pd.read_csv(filepath3)
df_2017['Year'] = 2017
df_2018 = pd.read_csv(filepath4)
df_2018['Year'] = 2018
df_2019 = pd.read_csv(filepath5)
df_2019['Year'] = 2019
df_2020 = pd.read_csv(filepath6)
df_2020['Year'] = 2020

df_2016.head()
```

```python
df_vertical = pd.DataFrame(columns = ['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)',
                                   'Social Support', 'Health Life Expectancy', 'Freedom','Trust (Government Corruption)',
                                   'Generosity', 'Year'])
years = [df_2015, df_2016, df_2017, df_2018, df_2019, df_2020]
for year in years:
    df_vertical = df_vertical.append(year[['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)',
                                   'Social Support', 'Health Life Expectancy', 'Freedom','Trust (Government Corruption)',
                                   'Generosity', 'Year']], ignore_index = 'True')
df_vertical
```

```python
df_vertical.to_csv("Vertical_Merge.csv", index=False)
```
