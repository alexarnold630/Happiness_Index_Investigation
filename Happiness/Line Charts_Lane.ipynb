## Line chart for the Top countries over the 5 years, iceland, denmark, norway
##### 

```python
# data science imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# linear modelling
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import scipy.stats as sts
```

```python
filepath1 = "Vertical_Merge.csv"
df = pd.read_csv(filepath1)
```

```python
df.head()
```

```python
df["Happiness Rank"].max()
```

```python
topdf = df.loc[df["Happiness Rank"] <= 5]
topdf.head()
```

```python
topdf["Country"].value_counts()
```

## Bottom countries 

```python
bottomdf = df.loc[df["Happiness Rank"] >= 141]
bottomdf["Country"].value_counts()
```

```python
rwanda = bottomdf.loc[bottomdf["Country"]=="Rwanda"]
rwanda
```

```python
afghanistan = bottomdf.loc[bottomdf["Country"]=="Afghanistan"]
afghanistan
```

```python
tanzania = bottomdf.loc[bottomdf["Country"]=="Tanzania"]
tanzania
```

```python
years = rwanda.Year
```

```python
rwandaHV = rwanda["Happiness Score"].values
afghanHV = afghanistan["Happiness Score"].values
tanzaniaHV = tanzania["Happiness Score"].values
```

```python
plt.figure(figsize=(10,6))

plt.plot(years, rwandaHV,color="#a0c4ff",label="Rwanda", linewidth=3)

plt.plot(years, afghanHV, color="#9bf6ff", label="Afghanistan", linewidth=3)

plt.plot(years, tanzaniaHV ,color="#bdb2ff", label="Tanzania", linewidth=3)

plt.legend()
plt.title("Bottom Countries Happiness Score over the Years", fontsize=18, fontweight="bold", fontname = "DejaVu Sans")
plt.xlabel("Years", fontsize=14, fontname = "DejaVu Sans")
plt.ylabel("Happiness Score", fontsize=14, fontname = "DejaVu Sans")

plt.savefig("Images/Bottom Countries Happiness Score over the Years.png")
plt.show()
```
