```python
 #data science - standard library
import pandas as pd
import numpy as np

# visualizations
import matplotlib.pyplot as plt
import seaborn as sns

from api_keys import g_key
import gmaps

import requests
import json
```

```python
# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
df_all = pd.read_csv("Vertical_Merge_LatLng_AllYears.csv")
df_all.head()
```

```python
 # Store latitude and longitude in locations
locations = df_all[["Latitude", "Longitude"]]
locations
```

```python
coords = locations
happy = df_all["Happiness Score"]

# Customize the size of the figure
figure_layout = {
    'width': '1200px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout, map_type = "SATELLITE")

# # Assign the marker layer to a variable
# markers = gmaps.marker_layer(coords, info_box_content=states)
# # Add the layer to the map
# fig.add_layer(markers)

heat = gmaps.heatmap_layer(coords, weights = happy, dissipating=False, max_intensity=100)
# Add the layer to the map
fig.add_layer(heat)

fig
```
