```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath1 = "Clean_2015.csv"
filepath2 = "Clean_2016.csv"
filepath3 = "Clean_2017.csv"
filepath4 = "Clean_2018.csv"
filepath5 = "Clean_2019.csv"
filepath6 = "Clean_2020.csv"

df_2015 = pd.read_csv(filepath1)
df_2016 = pd.read_csv(filepath2)
df_2017 = pd.read_csv(filepath3)
df_2018 = pd.read_csv(filepath4)
df_2019 = pd.read_csv(filepath5)
df_2020 = pd.read_csv(filepath6)

df_2020.head()
```

```python
df_combo = pd.merge(df_2015, df_2016, how="outer", on="Country", suffixes=["_2015", "_2016"]).reset_index(drop=True)
df_combo
```

```python
df_2017.columns
```

```python
df_combo.columns
```

```python
df_combo = pd.merge(df_combo, df_2017, how="outer", on="Country").rename(columns={"Happiness Rank": "Happiness Rank_2017", "Happiness Score":"Happiness Score_2017", "Economy (GDP per Capita)": "Economy (GDP per Capita)_2017", 
                                                                                 "Social Support": "Social Support_2017", "Health Life Expectancy": "Health Life Expectancy_2017", "Freedom": "Freedom_2017", "Trust (Government Corruption)": "Trust (Government Corruption)_2017", "Generosity": "Generosity_2017"})
df_combo
```

```python
df_combo = pd.merge(df_combo, df_2018, how="outer", on="Country").rename(columns={"Happiness Rank": "Happiness Rank_2018", "Happiness Score":"Happiness Score_2018", "Economy (GDP per Capita)": "Economy (GDP per Capita)_2018", 
                                                                                 "Social Support": "Social Support_2018", "Health Life Expectancy": "Health Life Expectancy_2018", "Freedom": "Freedom_2018", "Trust (Government Corruption)": "Trust (Government Corruption)_2018", "Generosity": "Generosity_2018"})
df_combo
```

```python
df_combo = pd.merge(df_combo, df_2019, how="outer", on="Country").rename(columns={"Happiness Rank": "Happiness Rank_2019", "Happiness Score":"Happiness Score_2019", "Economy (GDP per Capita)": "Economy (GDP per Capita)_2019", 
                                                                                 "Social Support": "Social Support_2019", "Health Life Expectancy": "Health Life Expectancy_2019", "Freedom": "Freedom_2019", "Trust (Government Corruption)": "Trust (Government Corruption)_2019", "Generosity": "Generosity_2019"})
df_combo
```

```python
df_combo = pd.merge(df_combo, df_2020, how="outer", on="Country").rename(columns={"Happiness Rank": "Happiness Rank_2020", "Happiness Score":"Happiness Score_2020", "Economy (GDP per Capita)": "Economy (GDP per Capita)_2020", 
                                                                                 "Social Support": "Social Support_2020", "Health Life Expectancy": "Health Life Expectancy_2020", "Freedom": "Freedom_2020", "Trust (Government Corruption)": "Trust (Government Corruption)_2020", "Generosity": "Generosity_2020"})
df_combo
```
